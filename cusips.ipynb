```python
import os
import pandas as pd
```

```python
csv_files = []
for root, dirs, files in os.walk("data/pdf13flist"):
    for file in files:
        if file.endswith(".csv"):
            csv_files.append(os.path.join(root, file))
csv_files = sorted(csv_files)
csv_files
```

```python
cusips_df = pd.DataFrame()
for path in csv_files:
    df = pd.read_csv(path)
    cusips_df = pd.concat([cusips_df, df], ignore_index=True)
cusips_full_df = cusips_df.copy()
# cusips_df.drop(columns=["run_date", "run_time", "year", "quarter"], inplace=True)
# cusips_df.dropna(subset=["cusip", "issuer_name"], inplace=True)
cusips_df.drop_duplicates(inplace=True)
cusips_df.info()
cusips_df
```
